```python
#Sales units , Turnover -> all should turn to integers as data types
df["Sales units"] =pd.to_numeric( df["Sales units"] , downcast='integer' , errors ='coerce')
df["Sales units"] =df["Sales units"].astype ('Int64')
df["Turnover"] =pd.to_numeric( df["Turnover"] , downcast='integer' , errors ='coerce')
df["Turnover"] =df["Turnover"].astype ('Int64')
#Customer , StoreID , Dept_ID , Time index  -> should be dropped
df.drop(columns=["Customer","StoreID","Dept_ID","Time index"], axis =1 , inplace = True)
```

```python
#remove any blank rows in the table(8 rows )
Index= df[df["MonthYear"]==" - - - - "].index
Index
df.drop(Index , axis = 0 , inplace=True)
```
